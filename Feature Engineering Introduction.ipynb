# **What is Feature Engineering**

Process of using domain knowledge to extract feactures from raw data.

These features can be used to improve the performance of ML algorithm.

Screenshot 2024-04-23 131156.png

**Feature Transformation** --> One feature / input column transform into another form so it will perform better during training.

**Feature Construction** -->  Not have some feature availabe directly, we will create them manually, because we thought that if we have these features, model will perform better.

**Feature Selection** --> We don't use all columns, Ignore irrelevent columns.

**Feature Extraction** --> From given features, Completely extract new features programatically
